# **6. Pembangunan Model Clustering**

## **a. Pembangunan Model Clustering**

Pada tahap ini, Anda membangun model clustering dengan memilih algoritma yang sesuai untuk mengelompokkan data berdasarkan kesamaan. Berikut adalah **rekomendasi** tahapannya.
1. Pilih algoritma clustering yang sesuai.
2. Latih model dengan data menggunakan algoritma tersebut.

```python
data_encoded = df.copy()

silhouette_scores = []
for k in range(2, 21):  # Mencoba 2 hingga 20 klaster
    kmeans = KMeans(n_clusters=k, random_state=0)
    clusters = kmeans.fit_predict(data_encoded)
    silhouette_scores.append(silhouette_score(data_encoded, clusters))

optimal_k = silhouette_scores.index(max(silhouette_scores)) + 2  # Karena indeks mulai dari 2
print(f"Optimal K: {optimal_k}")
```

```python
# Fungsi menjalankan KMeans and hitung silhouette score
def run_kmeans(X, n_clusters=optimal_k):
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    kmeans_clusters = kmeans.fit_predict(X)
    score = silhouette_score(X, kmeans_clusters)
    return kmeans_clusters, score
```

```python
kmeans_clusters, score_kmeans = run_kmeans(data_encoded)
print(f"Silhouette Score (KMeans Biasa): {score_kmeans:.4f}")
```

## **b. Evaluasi Model Clustering**

Untuk menentukan jumlah cluster yang optimal dalam model clustering, Anda dapat menggunakan metode Elbow atau Silhouette Score.

Metode ini membantu kita menemukan jumlah cluster yang memberikan pemisahan terbaik antar kelompok data, sehingga model yang dibangun dapat lebih efektif. Berikut adalah **rekomendasi** tahapannya.
1. Gunakan Silhouette Score dan Elbow Method untuk menentukan jumlah cluster optimal.
2. Hitung Silhouette Score sebagai ukuran kualitas cluster.

```python
# Mencari jumlah cluster optimal menggunakan Silhouette Score dan Elbow Method
silhouette_scores = []
inertia = []

# Coba clustering dengan jumlah cluster 2 hingga 20
for k in range(2, 21):
    kmeans = KMeans(n_clusters=k, random_state=0)
    clusters = kmeans.fit_predict(data_encoded)
    silhouette_scores.append(silhouette_score(data_encoded, clusters))
    inertia.append(kmeans.inertia_)

# Menentukan jumlah cluster optimal dengan Silhouette Score
optimal_k = silhouette_scores.index(max(silhouette_scores)) + 2  # Karena indeks mulai dari 2
print(f"Optimal K (Silhouette Score): {optimal_k}")

# Visualisasi Elbow Method (Inertia)
plt.figure(figsize=(10, 6))
plt.plot(range(2, 21), inertia, marker='o', label='Inertia')
plt.title('Elbow Method for Optimal K')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.grid(True)

# Menambahkan titik optimal pada Elbow Method
plt.scatter(optimal_k, inertia[optimal_k - 2], c='red', label=f'Optimal K = {optimal_k}', zorder=5)
plt.legend()
plt.show()

# Visualisasi Silhouette Score
plt.figure(figsize=(10, 6))
plt.plot(range(2, 21), silhouette_scores, marker='o', label='Silhouette Score')
plt.title('Silhouette Score for Different K')
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Score')
plt.grid(True)

# Menambahkan titik optimal pada Silhouette Score
plt.scatter(optimal_k, max(silhouette_scores), c='red', label=f'Optimal K = {optimal_k}', zorder=5)
plt.legend()
plt.show()
```
